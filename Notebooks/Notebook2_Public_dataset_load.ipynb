## Criterios de selección de datasets

Un dataset será considerado apto para el desarrollo inicial del IAI si cumple:

1. Contiene expresión de al menos 60% de los marcadores del catálogo.
2. Usa símbolos génicos o puede mapearse a símbolos génicos.
3. Tiene suficientes muestras para entrenamiento o validación.
4. Incluye metadatos que permitan definir grupos biológicos o clínicos.
5. Es descargable en formato tabular o procesable con herramientas estándar.
